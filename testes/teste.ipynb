```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
if not api_key:
    print("OPENAI_API_KEY não encontrada no .env")
else:
    print("API key found and looks good so far!")

openai = OpenAI(api_key=api_key)

messages = [
    {"role":"system", "content":"você é um assistente que tira dúvidas sobre a Universidade Estadual do Piauí - UESPI"},
    {"role":"user", "content":"qual o número máximo de faltas para não reprovar?"},
    {"role":"user", "content":"sou estudante do curso de computação e a carga horária da matéria é de 60h"}
]

response = openai.chat.completions.create(model = "gpt-5-nano", messages=messages)

def display_summary(response):
    summary = response.choices[0].message.content
    display(Markdown(summary))

display_summary(response)
```
